## 3) Artificial neuron — from math to code

A neuron computes:

$$
z = \sum_i w_i x_i + b
$$

Then applies an activation function:

$$
y = g(z)
$$

This section connects directly to the theory seen in class.


```scala
x = torch.tensor([1.0, -2.0, 3.0])
w = torch.tensor([0.2, 0.4, -0.1])
b = torch.tensor(0.1)

z = torch.sum(x * w) + b
z
```

### Activation functions

1. Implement **ReLU**
2. Implement **Sigmoid**
3. Apply both to `z` and compare the outputs

Which activation preserves negative values?
- Sigmoid preserves negatives in the sense that for negative z it outputs a positive value close to 0 (it does not clip to 0), while ReLU clips all negative inputs to 0.

```scala
# TODO
def relu(z):
    return torch.clamp(z, min=0.0)

def sigmoid(z):
    return 1 / (1 + torch.exp(-z))

y_relu = relu(z)
y_sigmoid = sigmoid(z)

y_relu, y_sigmoid
```

## 4) Autograd and gradients

PyTorch uses **automatic differentiation** to compute gradients
using the **chain rule** (backpropagation).


```scala
x = torch.tensor([1.0, 2.0, -1.0], requires_grad=True)
w = torch.tensor([0.5, -0.3, 0.8], requires_grad=True)
b = torch.tensor(0.2, requires_grad=True)

z = torch.sum(x * w) + b
loss = (z - 1.0) ** 2

loss.backward()

print("loss:", loss.item())
print("grad w:", w.grad)
print("grad b:", b.grad)
```

### 🔍 Conceptual question

- If `b.grad > 0`, should `b` increase or decrease after a gradient descent step?
Explain **why** in one sentence.

- If **b.grad > 0**, then **b should decrease** after a gradient descent step, because of gradient descent updates, so subtracting a positive gradient makes \( b \) smaller and reduces the loss.


## 5) Toy classification dataset

We create a **linearly separable** dataset.

Label rule:
- class = 1 if `x₁ + x₂ + x₃ > 0`
- else class = 0

This mimics a very simple classification problem.


```scala
# TODO: generate a dataset of size N=500 with 3 features
N = 500
X = torch.randn(N, 3)
y = (X.sum(dim=1, keepdim=True) > 0).float()  # shape (N, 1)

# TODO: split into train (80%) and validation (20%)
n_train = int(0.8 * N)
perm = torch.randperm(N)

train_idx = perm[:n_train]
val_idx = perm[n_train:]

X_train, y_train = X[train_idx], y[train_idx]
X_val, y_val = X[val_idx], y[val_idx]
```

## 6) Model definition

We define a small **MLP** (fully-connected network):

`3 → 16 → 8 → 1`

Activation: ReLU  
Output: raw logits (no sigmoid)


```scala
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(3, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, 1)
        )

    def forward(self, x):
        return self.net(x)

# TODO: create model and move it to the GPU
model = MLP().to(device)
```

###  parameters

1. Compute **by hand** the total number of parameters
2. Verify your answer using PyTorch


```scala
# TODO: count parameters with PyTorch
total_params = sum(p.numel() for p in model.parameters())
total_params
```

## 7) Training loop

You must complete the full training loop:
- forward pass
- loss computation
- backward pass
- optimizer step

Loss: `BCEWithLogitsLoss`
Optimizer: `SGD`


```scala
# TODO: move data to device
X_train_d = X_train.to(device)
y_train_d = y_train.to(device)
X_val_d = X_val.to(device)
y_val_d = y_val.to(device)

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

for epoch in range(20):
    model.train()
    optimizer.zero_grad()

    # TODO: forward
    logits = model(X_train_d)

    # TODO: loss
    loss = criterion(logits, y_train_d)

    # TODO: backward
    loss.backward()

    # TODO: update
    optimizer.step()

    if epoch % 5 == 0:
        print("Epoch", epoch, "| loss =", float(loss))
```

## 8) Evaluation

1. Apply `sigmoid` to the logits
2. Convert probabilities to predictions
3. Compute **accuracy** on the validation set


```scala
# TODO: evaluation
with torch.no_grad():
    logits = model(X_val_d)
    probs = torch.sigmoid(logits)
    preds = (probs >= 0.5).float()

accuracy = (preds == y_val_d).float().mean()
accuracy
```

### 9) Reflection questions

- **Why do we not apply sigmoid inside the model?**  
  Because we use `BCEWithLogitsLoss`, which expects raw logits and internally applies a numerically stable sigmoid.  
  Adding sigmoid in the model would apply it twice and can hurt training stability.

- **What would happen if we removed all ReLU activations?**  
  The network would become a purely linear model (a stack of linear layers is still linear).  
  It could only learn a single linear decision boundary, reducing its expressive power.

- **How does this toy problem relate to image classification?**  
  It’s the same idea: learn a function that maps input features to a class label using a loss and gradient descent.  
  For images, inputs are high-dimensional pixels, and we need models that exploit spatial structure (e.g., CNNs) to generalize well.
